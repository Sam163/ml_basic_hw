# Задача 1
Написать функцию, которая будет перводить снейк_кейс в КэмелКейс и наоборот. Функция сама определяет - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

Пример:
- otus_course -> OtusCourse
- PythonIsTheBest -> python_is_the_best

```python
def to_camel_case(snake_str: str) -> str:
    camel_str = ""
    word_start = True
    for char in snake_str:
        if word_start and char.isalpha():
            camel_str = camel_str + char.upper()
            word_start = False
            continue
        if char == "_":
            word_start = True
            continue
        camel_str = camel_str + char
    return camel_str

def to_snake_case(camel_str: str) -> str:
    snake_str = ""
    for char in camel_str:
        if not snake_str:
            snake_str = char.lower()
            continue
        if char.isupper():
            snake_str = snake_str + "_" + char.lower()
            continue
        snake_str = snake_str + char
    return snake_str
    
def change_case(s: str) -> str:
    if not s:
        return s
    if "_" not in s and s[0].isupper():
        return to_snake_case(s)
    return to_camel_case(s)

cases = {
          "otus_course": "OtusCourse",
          "PythonIsTheBest": "python_is_the_best",
          "python": "Python",
          "Python": "python",
          "case_some1": "CaseSome1",
          "CaseSome1": "case_some1",
          "case_7": "Case7",
        }

for case, expected_result in cases.items():
    result = change_case(case)
    if (result == expected_result):
        print(case, "->", result)
    else:
        print("Wrong result:", result, "for case:", case, "Expected:", expected_result)

x = input()
print(change_case(x))
```

# Задача 2
Написать функцию проверяющую валидность введенной даты.

Пример:
- 29.02.2000 -> True
- 29.02.2001 -> False
- 31.04.1962 -> False

```python
from typing import Optional

def get_number(num_str: str) -> Optional[int] :
    try:
        num = int(num_str)
    except:
        return None
    return num

def is_leap_year(year: int) -> bool:
    return (year % 4 == 0 and year % 100 != 0) or year % 400 == 0
    
def days_in_month(month: int, leap_year: bool) -> int: 
  if month in (1, 3, 5, 7, 8, 10, 12): 
    return 31 
  elif month == 2: 
    return 29 if leap_year else 28
  else: 
    return 30

def check_date(date: str, separator = ".", insignificant_zeros = True) -> bool:
    try:
        day, month, year = date.split(separator)
    except:
        return False

    if insignificant_zeros:
        if len(day) != 2 or len(month) !=2 or len(year) < 4:
            return False

    day = get_number(day)
    month = get_number(month)
    year = get_number(year)
    if day == None or month == None or year == None:
        return False

    if year < 0:
        return False
    if month <=0 or month > 12:
        return False

    leap = is_leap_year(year)
    days = days_in_month(month, leap)

    if day <= 0 or day > days:
        return False
    
    return True

cases = {
            "29.02.2000": True,
            "29.02.2001": False,
            "31.04.1962": False,
            "12.12.": False,
            "12..2321": False,
            "12.12": False,
            ".12.12": False,
            "12.x2.12": False,
            "12.00.12": False,
            "12.1.2001": False,
            "12.01.-1": False,
            "12.2.0001": False,
            "12.02.0001": True,
            "": False,
            "20": False
        }

for case, expected_result in cases.items():
    result = check_date(case)
    if (result == expected_result):
        print(case, "->", result)
    else:
        print("Wrong result:", result, "for case:", case, "Expected:", expected_result)
```

# Задача 3
Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

Пример:
- 17 -> True
- 20 -> False
- 23 -> True

```python
import math

def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    result = True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            result = False
            break
    return result

cases = {
            17: True,
            20: False,
            23: True,
            0: False,
            1: False,
            2: True,
            3: True,
            283: True,
            287: False,
        }

for case, expected_result in cases.items():
    result = is_prime(case)
    if (result == expected_result):
        print(case, "->", result)
    else:
        print("Wrong result:", result, "for case:", case, "Expected:", expected_result)
```

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. Также программа должна проверять, что имя и фамилия состоят только из символов и начинаются с большой буквы, если не с большой, то заменяет на большую, возраст должен быть числом от 18 до 60, ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным
Дополнительно написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
from typing import Optional

class DataBase:
    def __init__(self, age_limits: tuple[int, int] = (18, 60), key_size: int = 8):
        self._data = dict()
        self._age_limits = age_limits
        self._key_size = key_size

    def __str__(self):
        rows = []
        for key, record in self._data.items():
            rows.append("{} | {:<10} | {:<10} | {:<10}".format( self.key_to_str(key), *record))
        return '\n'.join(rows)
    
    @staticmethod
    def key_to_str(key: int) -> str:
        return str(key).rjust(8, '0')

    @staticmethod
    def _get_num(num_str: str) -> Optional[int] :
        try:
            num = int(num_str)
        except:
            return None
        return num

    @staticmethod
    def format_name(name: str) -> str:
        return name[0].upper() + name[1:].lower()

    def parse(self, record: str) -> Optional[tuple]:
        try:
           name, surname, age_str, key_str = record.split()
        except:
            print(f"Некорректная строка '{record}'")
            return None
            
        def check_name(name: str):
            if len(name) == 0:
                print(f"Пустое имя.")
                return False
            if not name.isalpha():
                print(f"Имя '{name}' должно содержать только буквы.")
                return False
            return True
        
        if not check_name(name) or not check_name(surname) :
            return None

        name = self.format_name(name)
        surname = self.format_name(surname)

        age = self._get_num(age_str)
        if not age:
            print(f"Некорректный возраст {age_str}")
            return None
        if age < self._age_limits[0] or age > self._age_limits[1]:
            print(f"Возраст должен быть в диапазоне: {self._age_limits}")
            return None
        print(key_str)
        key = self._get_num(key_str)
        if key is None:
            print(f"Некоректный ID: {key}")
            return None

        key_limit = pow(10, self._key_size)
        if key < 0 or key > key_limit:
            print(f"ID должен быть от 0 до: {key_limit}")
            return None
        
        return (name, surname, age, key)
    
    def append(self, name: str, surname: str, age: int, key: int) -> bool:
        if key in self._data:
            return False

        self._data[key] = (name, surname, age)
        return True

    def try_parse_and_append(self, record: str) -> bool:
        parsed = self.parse(record)
        if not parsed:
            return False
        if not self.append(*parsed):
            print(f"ID {parsed[-1]} уже существует")
            return False
        return True


def default_example():
    data = ["иван иванов 50 1",
            "Сергей Козлов 42 2",
            "Жанна Агузарова 62 4356",
            "Андрюша Молодец 4 3",
            "КаРаБаС бАрАбАс 53 3",
            "Арлекино Драматургович 20 1000000000",
            "Нулевой Попавший 20 0",
            "Минус Неопавший 20 -1",
            "Нецелый Неопавший 20 1.2",
            "Бенедикт Камбербэтч 48 23",
            "Ибрагим Кандибобер _ 22 21",
            "Айди Повторюшкин 21 23",
            "R2 D2 30 2343",
            "RДва D2 30 2343",
            "Ким  33 333",
            "Чин   Ын   33 333",
            ]
    db = DataBase()
    for record in data:
        print("Input:", record)
        db.try_parse_and_append(record)
    print("\nИтоговые данные:")
    print(db)

def read_input() -> DataBase:
    db = DataBase()
    while True:
        print("Введите данные:")
        record = input()
        if not record:
            break
        if not db.try_parse_and_append(record):
            print("Данные не добавлены")
    return db

# default_example()
db = read_input()
print(db)
```
